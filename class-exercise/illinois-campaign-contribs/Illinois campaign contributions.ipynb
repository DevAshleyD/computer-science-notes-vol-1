```python
contributions['other'] = 100 - contributions.Dem - contributions.Repub
```

```python
# Make a stacked barchart of Dem vs Rep contributions

contributions.sort_values(['Dem', 'Repub', 'other']).plot(x='County',
        y=['Dem', 'Repub', 'other'],
        kind='barh',
        stacked=True,
        figsize = (6,15),
        width=0.8,
        colors=['blue','red','lightgrey'],
        title = 'Campaign contribution breakdown')

plt.tight_layout()

plt.savefig('contributions.pdf')
```
