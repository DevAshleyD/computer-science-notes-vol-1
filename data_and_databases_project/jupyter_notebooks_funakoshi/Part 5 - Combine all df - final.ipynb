## Clean up the dataframe by getting rid of whitespaces with strip()

```python
merged['citizenship'] = merged.citizenship.str.strip()
merged['years'] = merged.years.str.strip()
merged['full_name'] = merged.full_name.str.strip()
merged['event'] = merged.event.str.strip()
```

```python
split_years = merged.years.str.split('-')
for year in split_years:
    try:
        if year[1] < year[0]:
            print(year)
    except:
        pass
```

```python
merged.to_csv('medal_info.csv', index=False)
```
