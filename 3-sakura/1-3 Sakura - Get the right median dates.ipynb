```python
import pandas as pd
import re
import numpy as np
from datetime import datetime


import math
import statistics

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['pdf.fonttype'] = 42

%matplotlib inline

pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 1000)
pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_csv('data/sakura_complete_just_dates.csv')
df
```

```python
start = df.loc[:,'city':'2010_start']
start.head(1)
```

```python
full_cols = list(df.columns[-19:])
full_cols.append('city')
full = df[full_cols]
full.head(1)
```

```python
complete = start.loc[:, 'city':'longitude']
complete.head(1)
```

```python
complete['start_date'] = start.median(axis=1)
complete['full_date'] = full.median(axis=1)
```

```python
complete['bloom_duration'] = complete.full_date - complete.start_date
complete
```

```python
complete.to_csv('data/sakura-dates-fixed.csv', index=False)
```
