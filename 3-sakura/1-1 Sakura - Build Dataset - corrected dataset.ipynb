# Scrape coordinates and English name of cities from Google

```python
from bs4 import BeautifulSoup
import requests
```

```python
names = pd.read_csv('data/location-name.csv')
cities = list(names['loc'])
```

## Try scraping the city name in English & coordinates for one city

```python
city = cities[10]

url = f'https://www.google.com/search?q={city}座標'
response = requests.get(url)
doc = BeautifulSoup(response.text)

city_eng = doc.find(class_='BNeawe').text.strip()
coord = doc.find(class_='BNeawe iBp4i AP7Wnd').text.strip()
print(city_eng)
print(coord)
```

## Scrape city name (English) and corodinates for all cities

```python
rows = []
errors = []

for city in cities:
    row = {}
    error = {}
    try:
        url = f'https://www.google.com/search?q={city}座標'
        response = requests.get(url)
        doc = BeautifulSoup(response.text)
        row['city'] = city
        row['city_eng'] = doc.find(class_='BNeawe').text.strip()
        row['coord'] = doc.find(class_='BNeawe iBp4i AP7Wnd').text.strip()
        rows.append(row)
    except:
        error['city'] = city
        error['city_eng'] = 'ERROR'
        error['coord'] = 'ERROR'
        print(city)
        errors.append(error)
```
